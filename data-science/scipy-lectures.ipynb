## Matplotlib: Plotting

### Sin/Cos Plotting

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(C)
plt.plot(S)

plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.figure(figsize=(8, 6), dpi=100)
plt.subplot(1, 1, 1)

plt.plot(X, C, color="blue", linewidth=1.0, linestyle="--")
plt.plot(X, S, color="pink", linewidth=4.5, linestyle="-")

plt.xlim(-4.0, 4.0)
plt.ylim(-1.0, 1.0)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])

plt.show()
```

### Add Axis

```python
plt.figure(figsize=(10, 6), dpi=100)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])

plt.plot(X, C**5*S, color="blue", linewidth=2.5, linestyle="-", label="C**5*S")
plt.plot(X, S**3*S**2, color="red",  linewidth=2.5, linestyle="-", label="S**3*S**2")

plt.legend(loc='upper left')

ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.show()
```

### Add Annotation Point

```python
plt.figure(figsize=(8, 6), dpi=100)
plt.subplot(1, 1, 1)

plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")
plt.plot(X, S, color="pink", linewidth=1.0, linestyle="-")

plt.xlim(-4.0, 4.0)
plt.ylim(-1.0, 1.0)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])

ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

t = 2*np.pi/3

plt.plot([t, t], [0, np.cos(t)], linewidth=2.0, linestyle="--", color="blue")
plt.scatter([t,], [np.cos(t)], 60, color="red")
plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$', xy=(t, np.cos(t)),
             xycoords='data', xytext=(-90, -50), textcoords='offset points',
             fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.show()
```
